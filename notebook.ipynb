For now, let's go ahead and convert the datetime column to an actual datetime object. We should also set some columns to the categorical type, and clean the column names just in case.

```python
df_police.columns = df_police.columns.str.strip()
df_police['Event Date/Time'] = pd.to_datetime(df_police['Event Date/Time'])
categoricals = ['ID Reference Number', 'Subject\'s race', 'Subject\'s gender', 'Ethnicity', 'District of occurrence']
df_police[categoricals] = df_police[categoricals].astype('category')
```

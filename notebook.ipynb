```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df_police = pd.read_csv('Police_Data.csv')
df_police.head()
```

```python
df_police=df_police.rename(columns={"Subject's race":"Race", "Subject's gender": "Gender", "Subject's age": "Age", "District of occurrence": "District"})
df_police
```

```python
#Race
plt.figure(figsize=(10, 6))
df_police["Race"].value_counts().plot(kind='barh')
plt.title('Distribution of Race')
plt.xlabel('Count')
plt.ylabel('Race')
plt.show()
```

```python
#Gender 
plt.figure(figsize=(6, 4))
df_police["Gender"].value_counts().plot(kind='bar')
plt.title('Distribution of Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
#Ethnicity 
plt.figure(figsize=(6, 4))
df_police['Ethnicity'].value_counts().plot(kind='bar')
plt.title('Distribution of Ethnicity')
plt.xlabel('Ethnicity')
plt.ylabel('Count')
plt.show()
```

```python
#Age 
plt.figure(figsize=(10, 6))
df_police["Age"].plot(kind='hist', bins=30, edgecolor='black')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
df_police['Event Date/Time'] = pd.to_datetime(df_police['Event Date/Time'])
df_police['Arrest Date'] = df_police['Event Date/Time'].dt.date
df_police['Arrest Year'] = df_police['Event Date/Time'].dt.year
df_police['Arrest Month'] = df_police['Event Date/Time'].dt.to_period('M')
```
